```python
import numpy as np
import pandas as pd
from pylab import *
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as datetime
import pandas_datareader as pdr
```

```python
# retreive dogecoin price data on specific data range
df = pdr.get_data_yahoo('DOGE-USD', start= '1-1-2019', end = '8-13-2022')
```

```python
df.head()
```

```python
df.columns
```

```python

data_per_year = df[df.columns].resample('W').sum()
data_per_year.head()
```

Dogecoin raising suddenly in 2021 on a seven month run.

```python
f, ax = plt.subplots(nrows=6, ncols=1, figsize=(15, 25))
for i, column in enumerate(data_per_year.columns):
    sns.lineplot(x=data_per_year.index, y=data_per_year[column], ax=ax[i], color='green')
    ax[i].set_title(f'{column}', fontsize=14)
    ax[i].set_ylabel(ylabel=column, fontsize=14)
    tight_layout()
```

Average exponential growth and crash between 2021 - 2022 for the daily price highs.

```python
df['High']['2021':'2022'].resample('D').mean().plot()
```

Average volume or amount of trading activity made daily.

```python
df['Volume']['2021':'2022'].resample('W').mean().plot()
```
